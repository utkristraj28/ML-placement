# Working and Cleaning

```python
place.head(10)
```

```python
gender_map = {'M':0,'F':1}
place['gender'] = place['gender'].map(gender_map)
```

```python
ssc_map = {'Others':1,'Central':0}
hsc_map = {'Others':1,'Central':0}
hscs_map = {'Commerce':0,'Science':1,'Arts':2}
degree_map = {'Sci&Tech':0,'Comm&Mgmt':1,'Others':2}
workex_map = {'Yes':1,'No':0}
special_map = {'Mkt&HR':0,'Mkt&Fin':1}
```

```python
place['ssc_b'] = place['ssc_b'].map(ssc_map)
place['hsc_b'] = place['hsc_b'].map(hsc_map)
place['hsc_s'] = place['hsc_s'].map(hscs_map)
place['degree_t'] = place['degree_t'].map(degree_map)
place['workex'] = place['workex'].map(workex_map)
place['specialisation'] = place['specialisation'].map(special_map)
```

```python
status_map = {'Not Placed':0,'Placed':1}
place['status'] = place['status'].map(status_map)
```

```python
place.head()
```

```python
place.info()
```

# Training Testing

```python
x = place.drop('status',axis=1)
y = place['status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,random_state=101)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
pred = model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(ytest,pred))
print('\n')
print(classification_report(ytest,pred))
```

# SVM

```python
from sklearn.svm import SVC
```

```python
svc  = SVC()
```

```python
svc.fit(xtrain,ytrain)
```

```python
pred_svc = svc.predict(xtest)
```

```python
print(confusion_matrix(ytest,pred_svc))
print('\n')
print(classification_report(ytest,pred_svc))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C':[0.1,1,10,100,1000],'gamma':[1,0.1,0.01,0.001,0.0001]}
```

```python
grid = GridSearchCV(SVC(),param_grid,verbose=3)
```

```python
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
pred_grid = grid.predict(xtest)
```

```python
print(confusion_matrix(ytest,pred_grid))
print('\n')
print(classification_report(ytest,pred_grid))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
#knn = KNeighborsClassifier(n_neighbors=)

error = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_knn = knn.predict(xtest)
    error.append(np.mean(pred_knn != ytest))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error,color='blue',ls='--',marker='o',markerfacecolor='red')
plt.xlabel('knn value')
plt.ylabel('error rate')
```

```python
knn2 = KNeighborsClassifier(n_neighbors=7)
```

```python
knn2.fit(xtrain,ytrain)
```

```python
pred_knn2 = knn2.predict(xtest)
```

```python
print(confusion_matrix(ytest,pred_knn2))
print('\n')
print(classification_report(ytest,pred_knn2))
```
